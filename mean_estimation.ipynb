# Mean Approximation

```python
import numpy as np
teta = 0
random_variable =[]
for i in range(100):
    random_variable.append(np.random.uniform(2,60))

n = len(random_variable)
```

```python
teta_l= []
for i in range(len(random_variable)):
    teta = teta + (1/(i+1)) * (random_variable[i]-teta)
    teta_l.append(teta)
    if teta > max(random_variable):
        teta = max(random_variable)
    elif teta<min(random_variable):
        teta = min(random_variable)
```

```python
teta
```

```python
z = np.mean(random_variable)
z
```

```python
df = pd.DataFrame({'teta_l':teta_l})
df['z']=z
```

```python
import plotly_express as px
import pandas as pd

my_plot = px.line(df, x=(np.linspace(1,k,100)),
                     y=[df['teta_l'],df['z']],
                     title="Robbin Monro: Mean approximation")


my_plot.show()
```
